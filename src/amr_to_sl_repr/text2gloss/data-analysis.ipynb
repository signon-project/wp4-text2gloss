```python
import json
from pathlib import Path
```

```python
pfin = Path(r"D:\corpora\sl\Corpus VGT\annotations.json")
data = json.loads(pfin.read_text(encoding="utf-8"))
```

```python
glosses = sorted([g["gloss"] for item in data for side in ("left", "right") for g in item["glosses"][side]])
glosses_uniq = sorted(set(glosses))
```
